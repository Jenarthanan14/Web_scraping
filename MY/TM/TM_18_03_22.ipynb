```python
#Loading Packages
import numpy as np
import re
import string
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import openpyxl
import time

import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#Creating path for data extraction
now = datetime.now()
extraction_date = now.date()
date = extraction_date.strftime("%Y%m%d")
path = 'C:/Users/JenarthananRajenthir/OneDrive - ADA Asia/Documents/Final_telco/Data/MY/TM/'+  date + '/'              
if not os.path.exists(path):
    os.makedirs(path)
    
Carrier = 'TM'
```

```python
################### Explore Prepaid #####################################
df_Prepaid = pd.DataFrame(columns=['Carrier','Package_Type', 'Package_Option', 'Package_Name','Data', 'Calls',
                                   'SMS', 'Benefits', 'Info', 'Validity', 'Price'])

Package_Type = 'Prepaid'
```

```python
# Extracting Prepaid Packages
url1 = 'https://home.unifi.com.my/personal/mobile/bebas'

html = requests.get(url1)
soup1 = BeautifulSoup(html.content, "html5lib")

Package_Option = 'Prepaid'
for i in range(3, 13):
    DS = soup1.find_all('div', {'class': 'item'})[i].text.split('\n                                         ')
    xl = [x.strip() for x in DS if x.strip()]
    
    Package_Name = xl[0]+' - '+xl[1]
    if i < 7:
        Data = xl[2]
        Validity = xl[1]
        Calls = None 
        SMS = None
        Info = None
        
        if len(xl) == 5:
            Price = xl[4]
            Benefits = None
            
        elif len(xl) == 6:
            Price = 'RM '+xl[5]
            Benefits = 'Initial Price- '+xl[4].replace('\xa0', '').replace('\u200b', '')
            
        else:
            continue
         
    
        df_Prepaid = df_Prepaid.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                        'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                        'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                       ignore_index = True)
        
    elif i < 11:
        Data = xl[0]
        Validity = None
        Calls = None 
        SMS = None
        Info = None
        
        if len(xl) == 4:
            Price = xl[3]
            Benefits = xl[2]
            
        elif len(xl) == 5:
            Price = xl[4]
            Benefits = xl[2]+' '+xl[3]
            
        else:
            continue
         
    
        df_Prepaid = df_Prepaid.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                        'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                        'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                       ignore_index = True)
        
    else:
        Data = None
        Validity = None
        Calls = xl[0] 
        SMS = None
        Info = None
        Price = xl[3]
        Benefits = None
            
         
        df_Prepaid = df_Prepaid.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                        'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                        'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                       ignore_index = True)
        

        
# Extracting Limited Edition Starter Pack
DS = soup1.find_all('div', {'class' : 'starterpack-box limited fadeUp'})[0].text.split('\n                    ')
xl = [x.strip() for x in DS if x.strip()]

Package_Name = soup1.find_all('h2')[6].text
Data = xl[0]
Validity = xl[3]
Calls = xl[1] 
SMS = xl[2]
Info = 'Limited Time Only'
Price = 'RM 10'
Benefits = None       


df_Prepaid = df_Prepaid.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                        'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                        'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                       ignore_index = True)
        
    

df_Prepaid.head(12)
```

```python
################### Explore Postpaid #####################################
df_Postpaid = pd.DataFrame(columns=['Carrier','Package_Type', 'Package_Option', 'Package_Name','Data', 'Calls',
                                    'SMS', 'Benefits', 'Info', 'Validity', 'Price'])

Package_Type = 'Postpaid'
```

```python
# Extracting Postpaid Packages
url2 = 'https://my.unifi.com.my/personal/mobile/postpaid/ultimate'

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--ignore-certificate-errors')
options.add_argument('--allow-running-insecure-content')
options.add_argument("--disable-extensions")
options.add_argument("--proxy-server='direct://'")
options.add_argument("--proxy-bypass-list=*")
options.add_argument("--start-maximized")
options.add_argument('--disable-gpu')
options.add_argument("--incognito")
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument("--disable-blink-features=AutomationControlled")
browser = webdriver.Chrome(executable_path='../../chromedriver.exe', options = options)
browser.get(url2)
browser.implicitly_wait(10)
html_source_code = browser.execute_script("return document.innerHTML;")
browser.set_page_load_timeout(10)

html = browser.page_source
soup2 = BeautifulSoup(html, "html5lib")
browser.implicitly_wait(10)

Package_Option = 'Postpaid'
for i in soup2.find_all('div', {'class': 'css-1pcq3dh'}):
    Package_Name = i.find_all('div')[1].text
    Data = i.find_all('span')[1].text
    Validity = i.find_all('div')[1].text.split('/')[1]
    Calls =  i.find_all('span')[3].text
    SMS = i.find_all('span')[5].text
    Info = i.find_all('div')[0].text
    Price = i.find_all('div')[1].text.split('/')[0]
    Benefits =  None      

    df_Postpaid = df_Postpaid.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                      'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                      'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                     ignore_index = True)
    
    
# Extracting Ultimate Plan (manual extraction)
Package_Name = 'Ultimate Plan'
Data = 'Unlimited'
Validity = 'Month'
Calls = 'Unlimited'
SMS = 'Unlimited'
Info = None
Price = 'RM 79'
Benefits = 'RM 59 for 12 months contract and unifi subscribers'

df_Postpaid = df_Postpaid.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                  'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                  'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                 ignore_index = True)
    
df_Postpaid.head(6)    
```

```python
################### Explore Broadband #####################################
df_Broadband = pd.DataFrame(columns=['Carrier','Package_Type', 'Package_Option', 'Package_Name','Data', 'Calls',
                                     'SMS', 'Benefits', 'Info', 'Validity', 'Price'])

Package_Type = 'Broadband'
```

```python
# Extracting Fiber Packages
url3 = 'https://home.unifi.com.my/anytimeanywhere'

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--ignore-certificate-errors')
options.add_argument('--allow-running-insecure-content')
options.add_argument("--disable-extensions")
options.add_argument("--proxy-server='direct://'")
options.add_argument("--proxy-bypass-list=*")
options.add_argument("--start-maximized")
options.add_argument('--disable-gpu')
options.add_argument("--incognito")
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument("--disable-blink-features=AutomationControlled")
browser = webdriver.Chrome(executable_path='../../chromedriver.exe', options = options)
browser.get(url3)
browser.implicitly_wait(10)
html_source_code = browser.execute_script("return document.innerHTML;")
browser.set_page_load_timeout(10)

html = browser.page_source
soup3 = BeautifulSoup(html, "html5lib")
browser.implicitly_wait(10)

Package_Option = 'Fiber'
for i in soup3.find_all('div', {'class': 'row plan-wrapper'})[1].find_all('div', {'class': 'col-md-4 col-sm-6'}):
    xl = [x.strip() for x in i.text.split('\n                    ') if x.strip()]
    
    Package_Name = xl[0]
    Data = xl[0]
    Validity = xl[6]
    Calls = None
    SMS = None
    Price = xl[7]
    Benefits = xl[1]
    
    if xl[8] == 'RM 0/mth':
        Info = None
        
    else:
        Info = 'Old Price '+xl[8]
        

    df_Broadband = df_Broadband.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                        'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                        'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                       ignore_index = True)
    
# Extracting Wireless Broadband 
url4 = 'https://home.unifi.com.my/personal/home/wireless-broadband'

html = requests.get(url4)
soup4 = BeautifulSoup(html.content, "html5lib")

Package_Option = 'Wireless'
Package_Name = soup4.find_all('div', {'class': 'col-md-6'})[0].find_all('p')[1].text
Data = soup4.find_all('div', {'class': 'col-md-3'})[0].find_all('p')[1].text
Validity = soup4.find_all('div', {'class': 'col-md-6'})[1].text.split('\n')[3].split(' ')[1].split('/')[1]
Calls = None
SMS = None
Price = soup4.find_all('div', {'class': 'col-md-6'})[1].text.split('\n')[3].split(' ')[1].split('/')[0]
Benefits = soup4.find_all('div', {'class': 'col-md-3'})[1].find_all('p')[1].text
Info = 'Old Price- '+soup4.find_all('div', {'class': 'col-md-6'})[1].text.split('\n')[3].split(' ')[0]

df_Broadband = df_Broadband.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                    'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                    'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                   ignore_index = True)
    
        
    
# Extracting Mesh Wi-fi
url5 = 'https://home.unifi.com.my/unifi-experience'

html = requests.get(url5)
soup5 = BeautifulSoup(html.content, "html5lib")

Package_Option = 'Mesh Wi-Fi'
for i in soup5.find_all('div', {'class': 'col-md-5'}):
    xl = [x.strip() for x in i.text.split('\n             ') if x.strip()]
    Calls = None
    SMS = None
    Data = None
    
    if len(xl) == 6:
        Package_Name = xl[1]
        Validity = xl[3]
        Price = xl[2]
        Benefits = xl[5].split('\xa0')[0]
        Info = xl[0]
        
        df_Broadband = df_Broadband.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                            'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                            'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                           ignore_index = True)
        
    else:
        Package_Name = xl[2]
        Validity = xl[4]
        Price = xl[3]
        Benefits = xl[5]
        Info = xl[0]
        
        df_Broadband = df_Broadband.append({'Carrier': Carrier,'Package_Type' : Package_Type,'Package_Option' : Package_Option,
                                            'Package_Name' : Package_Name,'Data' : Data, 'Calls' : Calls, 'SMS' : SMS, 
                                            'Benefits' : Benefits, 'Validity' : Validity, 'Price' : Price, 'Info' : Info },
                                           ignore_index = True)

df_Broadband.head(12)
```
