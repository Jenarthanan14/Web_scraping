```python
#Loading Packages
import os
#import csv
import requests
from bs4 import BeautifulSoup
#import sys
import pandas as pd
#import re
from datetime import datetime
import time
now = datetime.now()
date = now.date()
```

```python
#Creating path for data extraction
now = datetime.now()
extraction_date = now.date()
date = extraction_date.strftime("%Y%m%d")
path = 'Mcalls/'+  date + '/'              
if not os.path.exists(path):
    os.makedirs(path)  
```

```python
#Extracting Devices Data
url_dev = 'https://mcalls.asia/ePrihatin/apply/prihatin-package-index'

html = requests.get(url_dev)
soup_dev = BeautifulSoup(html.content, "html5lib")
#print(soup_dev.prettify())             # visual representation of the parse tree
category_list = list(set(soup_dev.find_all('div', attrs = {'class':'captiontitle-wrap prihatin_cat_title text-center'})))



df_devices = pd.DataFrame(columns=['Category_Name', 'Category_Detail','Extraction_Date','Plan_Name', 'Plan_Info', 'Phone',
                                   'Price','Phone_Specification', 'Recommended_Retail_Price' ,'Upfront_Payment', 
                                   'Prepaid_Plan_Info_1','Prepaid_Plan_Info_2', 'Prepaid_Plan_Info_3','Sale_Status'])

#For loop done with 2 categories
for i in range(len(category_list)):
    if i == 0:
        for j in range(0,2,1):
            Phone_tab =soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('ul')[0].find_all('li')
            Plan_tab =soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('ul')[1].find_all('li') 
                
            Cat_Name = soup_dev.find_all('div', attrs = {'class':'captiontitle-wrap prihatin_cat_title text-center'})[i].find_all('p')[0].text
            Cat_Detail = soup_dev.find_all('div', attrs = {'class':'captiontitle-wrap prihatin_cat_title text-center'})[i].find_all('h3')[0].text
            Plan_Name = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('h4')[0].text
            Plan_Info = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('h4')[1].text
            Phone  = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('div', attrs = {'class':'plan-desc'})[0].find_all('p')[0].text
            Price = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('div', attrs = {'class':'plan-price'})[0].find_all('span')[0].text
            P1 = Plan_tab[0].text
            P2 = Plan_tab[1].text
            P3 = Plan_tab[2].text
            Sale_Stat = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('div', attrs = {'class':'plan'})[0].find_all('a', attrs = {'class':'btn btn-gradient'})[0].text
            Phone_Spec = Phone_tab[0].text
            Rec_Retail_Price = Phone_tab[1].text
            if len(Phone_tab) == 3:
                Upfront_Payment = Phone_tab[2].text
            else :
                Upfront_Payment = None
                
            df_devices = df_devices.append({'Category_Name': Cat_Name, 'Category_Detail':Cat_Detail,'Extraction_Date': extraction_date,
                                            'Plan_Name': Plan_Name, 'Plan_Info':Plan_Info, 'Phone': Phone, 'Price':Price,
                                            'Phone_Specification': Phone_Spec, 'Recommended_Retail_Price':Rec_Retail_Price ,
                                            'Upfront_Payment': Upfront_Payment, 'Prepaid_Plan_Info_1': P1,
                                            'Prepaid_Plan_Info_2': P2, 'Prepaid_Plan_Info_3': P3, 'Sale_Status': Sale_Stat}, 
                                           ignore_index=True)

    if i == 1:
        for j in range(2,4,1):
            Phone_tab =soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('ul')[0].find_all('li')
            Plan_tab =soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('ul')[1].find_all('li')
            
            Cat_Name = soup_dev.find_all('div', attrs = {'class':'captiontitle-wrap prihatin_cat_title text-center'})[i].find_all('p')[0].text
            Cat_Detail = soup_dev.find_all('div', attrs = {'class':'captiontitle-wrap prihatin_cat_title text-center'})[i].find_all('h3')[0].text
            Plan_Name = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('h4')[0].text
            Plan_Info = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('h4')[1].text
            Phone  = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('div', attrs = {'class':'plan-desc'})[0].find_all('p')[0].text
            Price = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('div', attrs = {'class':'plan-price'})[0].find_all('span')[0].text
            P1 = Plan_tab[0].text
            P2 = Plan_tab[1].text
            P3 = Plan_tab[2].text
            Sale_Stat = soup_dev.find_all('div', attrs = {'class':'col-md-4'})[j].find_all('div', attrs = {'class':'plan'})[0].find_all('a', attrs = {'class':'btn btn-gradient'})[0].text
            Phone_Spec = Phone_tab[0].text
            Rec_Retail_Price = Phone_tab[1].text
            if len(Phone_tab) == 3:
                Upfront_Payment = Phone_tab[2].text
            else :
                Upfront_Payment = None
                
            df_devices = df_devices.append({'Category_Name': Cat_Name, 'Category_Detail':Cat_Detail,'Extraction_Date': extraction_date,
                                            'Plan_Name': Plan_Name, 'Plan_Info':Plan_Info, 'Phone': Phone, 'Price':Price,
                                            'Phone_Specification': Phone_Spec, 'Recommended_Retail_Price':Rec_Retail_Price ,
                                            'Upfront_Payment': Upfront_Payment, 'Prepaid_Plan_Info_1': P1,
                                            'Prepaid_Plan_Info_2': P2, 'Prepaid_Plan_Info_3': P3, 'Sale_Status': Sale_Stat},
                                           ignore_index=True)

path ='C:/Users/JenarthananRajenthir/OneDrive - ADA Asia/Documents/Final_telco/Data/MY/mCalls/'+date + '/'
df_devices.to_csv(path+date+'_devices.csv', index=False)
print('Successfully Mcalls Device Details Collected!')
df_devices.head()
```
